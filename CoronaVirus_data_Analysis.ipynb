## Analysis of Covid  data before and after the locked down phase

In India from 21st of March locked down started and we are assuming that there will be 14 days time boundary after that a patient will  know he/she has corona virus or not.

```
statewise_data=pd.read_csv('https://api.covid19india.org/csv/latest/state_wise_daily.csv')
```

```
'''Statewise data of each day starting from {} to {}'''.format(statewise_data.iloc[0]['Date'],statewise_data.iloc[-1]['Date'])
```

```
statewise_data.head()
```

```
statewise_data['Date']=pd.to_datetime(statewise_data['Date'])
```

### Melt Function
Melt function is use to convert dataframe. It fix up any columns and associate it with all columns to create a new structured dataframe.



```

s1=statewise_data.melt(id_vars=['Date','Status'])
```

```
s1.head()
```

## State wise analysis

From the data we see that total number of cases is 78055.

Below the statewise confirmed data are given.

```
s1.query('Status=="Confirmed"').groupby(['variable']).sum().sort_values('value', ascending=False).iloc[1:,:]
```

```
ctab=pd.crosstab(s1['variable'],s1['Status'],s1['value'],aggfunc='sum')
```

```
ctab=ctab.reset_index()
```

```
ctab['Mortality']=ctab['Deceased']/ctab['Confirmed']
```

```
ctab.sort_values(['Mortality'], ascending=False)
```

From the above table we can conclude that  west bengal has highest number of mortality rate ~9%.


The plot below represent distribution of confirmed cases of statewise.

```
import plotly.express as px
fig=px.bar(ctab.query('variable!="TT"'), x='variable', y='Confirmed',hover_data=['Deceased','Mortality'])
fig.show()
```

```
fig2=px.bar(ctab.query('variable!="TT"'), x='variable', y='Deceased',hover_data=['Confirmed','Mortality'])
fig2.show()
```
